```python
import turicreate as tc
import pandas as pd
```

```python
df = pd.read_csv("gs://{Bucket Name}/blog/blog_data.csv")
df["clientid"] = df['clientid'].apply(lambda x: x.lstrip("'")) 
df = df.groupby(['clientid', 'blog']).agg({'rating':'sum'}).reset_index()
actions = tc.SFrame(df)
```

```python
training_data, validation_data = tc.recommender.util.random_split_by_user(actions, 'clientid', 'blog')
model = tc.recommender.item_similarity_recommender.create(training_data, 'clientid', 'blog', target="rating")
```

```python
results = model.recommend()
```

```python
clientid = "1001061588.1632666126"
visitied_blogs = df[df['clientid'] == clientid]['blog'].values.tolist()
predicted_blogs = results[
    results['clientid'] == clientid
]['blog']

print("Visited Blogs:\n{}\n".format("\n".join(visitied_blogs)))
print("Predicted Blogs:\n{}".format("\n".join(predicted_blogs)))
```
